```
### MK Code V2 ###

UN="dg@assiduousinvest.com"
PW="VAJpmdg0678"
headers = {'Content-Type': 'application/json', 'Accept': 'application/json'}
meta_datapoints=[
    ['Revenue', 190],
    ['Gross Profit', 194],
    ['Operating Profit', 204]]
meta_periods_actuals=[
    '3QCY-2017',
    '4QCY-2017',
    '1QCY-2018',
    '2QCY-2018',
    '3QCY-2018',
    '4QCY-2018',
    '1QCY-2019',
    '2QCY-2019',
    '3QCY-2019',
    '4QCY-2019',
    '1QCY-2020',
    '2QCY-2020',
]
meta_tickers= R3000_BBERG_Tickers

import requests
import pandas as pd

### It's fine for now to hard code your pw, but better practice is to just prompt it when you run the code
# def set_pw():
#     global PW
#     PW = input("Password for " + UN + ":")

def auth():
    url = 'https://app.visiblealpha.com/auth/'
    payload = {'username': UN, 'password': PW}
    response = requests.post(url, json = payload, headers = headers)
    if response.status_code == requests.codes.ok:
        t = response.json()['jwt'] 
        return t
    else:
        return 0
    
def get_actuals(tkr, pd, dp):
    url2 = 'https://app.visiblealpha.com/api2/' 
    param= 'getparam?param=' + str(dp) + '&'
    cut = 'cut=SD&' # options are CD & SD 
    ticker = 'ticker=' + tkr + ' US' + '&'
    tickertype='tickertype=B'
    period ='period=' + pd + '&'
    source = 'source=VA Actuals&' #options are "Consensus" “VA Actuals”, “Filings”
    revision = ''
    response = requests.get(url2+param+period+cut+ticker+source+revision+tickertype,headers=headers)
    item = response.json()
    return item

def get_consensus(tkr, pd, dt, dp):
    url2 = 'https://app.visiblealpha.com/api2/' 
    param= 'getparam?param=' + str(dp) + '&'
    cut = 'cut=SD&' # options are CD & SD 
    ticker = 'ticker=' + tkr + ' US' + '&'
    tickertype='tickertype=B'
    period ='period=' + pd + '&'
    source = 'source=Consensus&' #options are "Consensus" “VA Actuals”, “Filings”
    revision = 'revision=' + dt +'&'
    statistics = 'statistics=std,brokercount'
    response = requests.get(url2+param+period+cut+ticker+source+revision+statistics+tickertype,headers=headers)
    item = response.json()
    return item
        
#set_pw()
token = auth() 
headers['Authorization'] = 'Bearer '+ token

actuals_list = []
consensus_list = []

def next_quarter(this_period):
    y = int(this_period[-4:])
    q = int(this_period[0])
    if q == 4:
        q_new = 1
        y_new = y + 1
    else:
        q_new = q + 1
        y_new = y
    return str(q_new) + 'QCY-' + str(y_new)

for p in meta_periods_actuals:
    for t in meta_tickers:
        print(p,t)
        for d in meta_datapoints:
            d_name = d[0]
            d_code = d[1]
            # Get the actuals data for this period/ticker
            dict_actuals = { 'ticker': t, 'period': p, 'measure': d_name }  
            this_data = get_actuals( t, p, d_code )
            if( this_data['status']=='error' or len(this_data['data'])==0 ):
              print('VA API failed for ticker = ' + t)
              continue
            this_period_date = this_data['data'][0]['period_date']
            this_period_date = pd.Timestamp(this_period_date)
            dict_actuals['period_end_date'] = this_period_date
            dict_actuals['value'] = this_data['data'][0]['data'][0]['value']
            actuals_list.append( dict_actuals )
            # Now grab consensus for the same ticker/asOfDate for next 4 quarters
            p_cons = next_quarter(p)
            this_asOf_date = (this_period_date - pd.Timedelta(days=30)).strftime('%Y%m%d')
            i = 1
            while i <= 4:
                dict_consensus = { 'ticker': t, 'asOfDate': this_asOf_date, 'period': p_cons, 'measure': d_name }              
                this_cons_data = get_consensus(t, p_cons, this_asOf_date, d_code)
                if( this_cons_data['status']=='error' or len(this_cons_data['data'])==0 ):
                  print('VA API failed for ticker = ' + t)
                  i += 1
                  continue
                this_cons_period_date = this_cons_data['data'][0]['period_date']
                this_cons_period_date = pd.Timestamp(this_cons_period_date)
                dict_consensus['period_end_date'] = this_cons_period_date.strftime('%Y%m%d')
                this_cons_forecast_flag = this_cons_data['data'][0]['forecast']
                dict_consensus['forecast_flag'] = this_cons_forecast_flag
                this_cons_data = pd.json_normalize(this_cons_data, ['data','data'])
                dict_consensus['mean'] = this_cons_data['value'].item()
                if 'consensus_statistics.std' in this_cons_data.columns:
                  dict_consensus['std'] = this_cons_data['consensus_statistics.std'].item()
                else:
                  dict_consensus['std'] = ''
                if 'consensus_statistics.broker_count' in this_cons_data.columns:
                  dict_consensus['broker_count'] = this_cons_data['consensus_statistics.broker_count'].item()
                else:
                  dict_consensus['broker_count'] = ''
                consensus_list.append(dict_consensus)
                p_cons = next_quarter(p_cons)
                i += 1
# Get "synthetic actuals"
for t in meta_tickers:
  print( 'Synthetic Actuals for ' + t)
  p_cons = '4QCY-2019'
  this_asOf_date = '20200229'
  i = 1
  while i <= 6:
      dict_consensus = { 'ticker': t, 'asOfDate': this_asOf_date, 'period': p_cons, 'measure': d_name }              
      this_cons_data = get_consensus(t, p_cons, this_asOf_date, d_code)
      if( this_cons_data['status']=='error' or len(this_cons_data['data'])==0 ):
        print('VA API failed for ticker = ' + t)
        i += 1
        continue
      this_cons_period_date = this_cons_data['data'][0]['period_date']
      this_cons_period_date = pd.Timestamp(this_cons_period_date)
      dict_consensus['period_end_date'] = this_cons_period_date.strftime('%Y%m%d')
      this_cons_forecast_flag = this_cons_data['data'][0]['forecast']
      dict_consensus['forecast_flag'] = this_cons_forecast_flag
      this_cons_data = pd.json_normalize(this_cons_data, ['data','data'])
      dict_consensus['mean'] = this_cons_data['value'].item()
      if 'consensus_statistics.std' in this_cons_data.columns:
        dict_consensus['std'] = this_cons_data['consensus_statistics.std'].item()
      else:
        dict_consensus['std'] = ''
      if 'consensus_statistics.broker_count' in this_cons_data.columns:
        dict_consensus['broker_count'] = this_cons_data['consensus_statistics.broker_count'].item()
      else:
        dict_consensus['broker_count'] = ''
      consensus_list.append(dict_consensus)
      p_cons = next_quarter(p_cons)
      i += 1
actuals_data = pd.DataFrame( actuals_list )
consensus_data = pd.DataFrame( consensus_list )
```
